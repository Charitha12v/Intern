```python
import pandas as pd
import numpy as np
import seaborn as sns
data=pd.read_csv("/content/Netflix.csv",na_values=["Not Given","########"])
data
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.columns
print('Missing values of each column: \n',len(data) - data.count())
```

```python
data.duplicated().value_counts()
```

```python
data['release_year'].hist()
```

```python
q1= data['release_year'].quantile(0.25)
q3= data['release_year'].quantile(0.75)
iqr= q3-q1
print("iqr=", iqr)

lb=q1-1.5*iqr
ub=q3+1.5*iqr
print("lower bound=", lb)
print("upper bound=", ub)
```

```python
values_less_than_lb = len(data[data['release_year'] <= lb])
print("Number of values less than lb:", values_less_than_lb)
```

```python
missing = ['release_year']
for i in missing:
 data.loc[data.loc[:,i].isnull(),i]=data.loc[:,i].mean()
```
