```python
import pandas as pd
from sklearn.linear_model import LinearRegression
train_data=pd.read_html("https://docs.google.com/spreadsheets/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pubhtml")
test_data=pd.read_html("https://docs.google.com/spreadsheets/d/e/2PACX-1vRyvZ7lknwiSghK9aen1SaTEYoN3JS40rrGLpcyrsVZy1tB2T4gn6Y3-cdzPUFCPMmmqREWefW3kl4_/pubhtml")
set1=train_data[0]
set1.to_csv('train.csv')
train_data=pd.read_csv("train.csv")
train_data
trains=train_data.drop("Unnamed: 0",axis=1)
trains.columns=trains.iloc[0]
train_data=trains.drop(0)
train_data
```

```python
set2=test_data[0]
set2.to_csv("test.csv")
test_data=pd.read_csv("test.csv")
test_data
tests=test_data.drop("Unnamed: 0",axis=1)
tests.columns=tests.iloc[0]
test_data=tests.drop(0)
test_data
```

```python
train_dataset=train_data.dropna(subset=['y'])
train_dataset
```

```python
x_train=train_dataset[['x']].astype(float)
y_train=train_dataset[['y']].astype(float)
x_test=test_data[['x']].astype(float)
y_test=test_data[['y']].astype(float)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
yhat=model.predict(x_test)
yhat
```

```python
y_test=y_test.squeeze()
yhat=yhat.squeeze()
```

```python
show=pd.DataFrame({"Test value":y_test, "Predicted Value":yhat})
show
```
